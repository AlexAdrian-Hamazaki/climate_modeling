```python
import netCDF4 as nc

# Open the NetCDF file
dataset= nc.Dataset(path)
# Display dimensions
print("Dimensions:")
for dim_name, dimension in dataset.dimensions.items():
    print(f"{dim_name} = {len(dimension)}")

# Display variables
print("\nVariables:")
for var_name, variable in dataset.variables.items():
    print(f"{var_name}({', '.join(variable.dimensions)})")
    for attr in variable.ncattrs():
        print(f"    {attr} = {variable.getncattr(attr)}")


# Close the dataset
dataset.close()
```

```python
ds= nc.Dataset(path)

lat = ds.variables['lat']
lon = ds.variables['lon']
fwi = ds.variables['FWI']
```

```python
lo_lat = [i.data.item() for i in lat]
lo_lon = [i.data.item() for i in lon]
```

```python
lo_lat[0]
```

```python
lo_lon[0]
```

```python
lo_fwi_lon = []
for lon in lo_lon:
    fwi_lon = fwi[:,lon]
    print(fwi_lon)
    lo_fwi_lon.append(fwi_lon)
```

```python
fwi_df = pd.DataFrame(lo_fwi_lon)
fwi_df.columns  = lo_lat
fwi_df.index = lo_lon
fwi_df = fwi_df.reset_index().melt(id_vars = 'index', var_name = 'lat', value_name='fwi')
fwi_df.rename(columns={'index':'lon'},inplace=True)
```

```python
fwi_df.to_csv('fwi.csv')
```
