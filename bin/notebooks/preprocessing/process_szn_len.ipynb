```python
import netCDF4 as nc

# Open the NetCDF file
dataset= nc.Dataset(path)
# Display dimensions
print("Dimensions:")
for dim_name, dimension in dataset.dimensions.items():
    print(f"{dim_name} = {len(dimension)}")

# Display variables
print("\nVariables:")
for var_name, variable in dataset.variables.items():
    print(f"{var_name}({', '.join(variable.dimensions)})")
    for attr in variable.ncattrs():
        print(f"    {attr} = {variable.getncattr(attr)}")


# Close the dataset
dataset.close()
```

```python
ds= nc.Dataset(path)

lat = ds.variables['lat']
lon = ds.variables['lon']
szn_len = ds.variables['fire_season']

```

```python
szn_len.dimensions
```

```python
lo_lat = [i.data.item() for i in lat]
lo_lon = [i.data.item() for i in lon]
```

```python
lo_szn_len = []
for lon in lo_lon:
    lon_szn_len = szn_len[:,lon]
    lo_szn_len.append(lon_szn_len.data)
sznlen_df = pd.DataFrame(lo_szn_len)
sznlen_df.columns = lo_lat
sznlen_df.index = lo_lon
sznlen_df = sznlen_df.reset_index().melt(id_vars = 'index', var_name = 'lat', value_name='szn_len')
sznlen_df.rename(columns={'index':'lon'},inplace=True)
sznlen_df
```

```python
sznlen_df.to_csv('szn_len.csv')
```
